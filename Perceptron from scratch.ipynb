```python
from sklearn.datasets import make_classification
import numpy as np
X,y=make_classification(n_samples=100,n_features=2,hypercube=False,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,class_sep=10)
print(X.shape)
print(y.shape)
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],cmap="winter",c=y,s=100)
```

```python
def perceptron(X,y):
  X=np.insert(X,0,1,axis=1) # insert(data,postion where new column added,value inside the new column,column)
  weights=np.ones(X.shape[1])
  lr=0.01

  for i in range(1000):
    j=np.random.randint(0,100)
    y_hat=step(np.dot(X[j],weights)) # equation of line
    weights=weights+lr*(y[j]-y_hat)*X[j]
  return weights


```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
weights=np.ones(X.shape[1])
weights
```

```python
k=perceptron(X,y)
```

```python
intercept=k[0]
slope=k[1:]
print(intercept)
print(slope)
```

```python
m=-slope[0]/slope[1]
c=-intercept/slope[1]
print(m)
print(c)
```

```python
X_input=np.linspace(3,-3,1000)
print(X_input)
y_input=m*X_input +c
```

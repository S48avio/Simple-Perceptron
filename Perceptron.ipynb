```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("placement.csv")
```

```python
df
```

```python
plt.scatter(df['cgpa'],df['resume_score'],c=df['placed'])
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import Perceptron
```

```python
pf=Perceptron()
pf.fit(x,y)
```

```python
pf.coef_  # here w1 weight is 40.26 and w2 is -36
```

```python
pf.intercept_   # tells the intercept value or the biase value
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values,y.values,clf=pf,legend=2)
```
